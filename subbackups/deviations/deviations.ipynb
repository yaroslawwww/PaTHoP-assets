```python
import matplotlib.pyplot as plt
import pandas as pd

# Чтение данных из файла
df = pd.read_csv('deviation_results.txt', header=None)
df[0] = df[0]+28
df[5] = df[5]*100
# Находим строку, где последний элемент равен 5000
special_row = df[df[6] == 5000]

# Исключаем специальную строку из основного набора данных
df_without_special = df[df[6] != 5000]

# Сортировка по первому столбцу
df_sorted = df_without_special.sort_values(by=0)
# Создание графиков
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 12))

# График зависимости 4-го столбца от 1-го
ax1.plot(df_sorted[0], df_sorted[3], 'b-', label='Суммарный размер ряда = 10000')
if not special_row.empty:
    # Рисуем горизонтальную пунктирную линию через значение y специальной точки
    ax1.axhline(y=special_row[3].values[0], color='red', linestyle='--', alpha=0.7, linewidth=1)
    # Добавляем точку (менее жирную)
    ax1.scatter(special_row[0], special_row[3], color='red', s=50, alpha=0.8,
                label='Последний элемент = 5000', zorder=5)
ax1.set_xlabel('Значение параметра r')
ax1.set_ylabel('RMSE')
ax1.set_title('Зависимость RMSE от r')
ax1.grid(True)
ax1.legend()

# График зависимости 5-го столбца от 1-го
ax2.plot(df_sorted[0], df_sorted[4], 'r-', label='Суммарный размер ряда = 10000')
if not special_row.empty:
    # Рисуем горизонтальную пунктирную линию через значение y специальной точки
    ax2.axhline(y=special_row[4].values[0], color='red', linestyle='--', alpha=0.7, linewidth=1)
    # Добавляем точку (менее жирную)
    ax2.scatter(special_row[0], special_row[4], color='red', s=50, alpha=0.8,
                label='Последний элемент = 5000', zorder=5)
ax2.set_xlabel('Значение параметра r')
ax2.set_ylabel('NP точки(доля)')
ax2.set_title('Зависимость количества np точек от r')
ax2.grid(True)
ax2.legend()

# График зависимости 6-го столбца от 1-го
ax3.plot(df_sorted[0], df_sorted[5], 'g-', label='Суммарный размер ряда = 10000')
if not special_row.empty:
    # Рисуем горизонтальную пунктирную линию через значение y специальной точки
    ax3.axhline(y=special_row[5].values[0], color='red', linestyle='--', alpha=0.7, linewidth=1)
    # Добавляем точку (менее жирную)
    ax3.scatter(special_row[0], special_row[5], color='red', s=50, alpha=0.8,
                label='Последний элемент = 5000', zorder=5)
ax3.set_xlabel('Значение параметра r')
ax3.set_ylabel('MAPE(%)')
ax3.set_title('Зависимость MAPE от r')
ax3.grid(True)
ax3.legend()

plt.tight_layout()
plt.show()
```
