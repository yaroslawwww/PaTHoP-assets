```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных
df = pd.read_csv('share_experiment.txt', header=None)

# Группировка по первому столбцу
grouped = df.groupby(0)

# Создание графиков для каждой группы
for name, group in grouped:
    # Сортировка по второму столбцу для упорядоченного отображения
    group_sorted = group.sort_values(by=1)

    # Построение трех графиков
    fig, axes = plt.subplots(3, 1, figsize=(10, 12))
    fig.suptitle(f'r = {name}')

    # График для 4-го столбца
    axes[0].plot(group_sorted[1], group_sorted[3], 'o-', label='RMSE')
    axes[0].set_ylabel('RMSE')
    axes[0].legend()

    # График для 5-го столбца
    axes[1].plot(group_sorted[1], group_sorted[4], 's-', label='NP точки(доля)')
    axes[1].set_ylabel('NP точки(доля)')
    axes[1].legend()

    # График для 6-го столбца
    axes[2].plot(group_sorted[1], group_sorted[5]*100, '^-', label='MAPE(%)')
    axes[2].set_xlabel('Второй столбец')
    axes[2].set_ylabel('MAPE(%)')
    axes[2].legend()

    # Сохранение графиков в файл
    plt.savefig(f'r={name}.png')
    plt.close()
```
