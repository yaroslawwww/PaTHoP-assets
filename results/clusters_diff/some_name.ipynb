```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных из файла (разделитель - любое количество пробелов)
data = pd.read_csv('ari_results.txt', delim_whitespace=True)

# Сортировка данных по r_value для правильного отображения графика
data_sorted = data.sort_values('r_value')

# Построение графика
plt.figure(figsize=(10, 6))
plt.plot(data_sorted['r_value'], data_sorted['ARI'], marker='o', linestyle='-', color='b')
plt.xlabel('r_value')
plt.ylabel('ARI')
plt.title('Зависимость ARI от r_value')
plt.grid(True)
plt.show()
```
