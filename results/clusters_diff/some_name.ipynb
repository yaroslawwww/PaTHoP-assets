```python
import matplotlib.pyplot as plt
import re

# Чтение данных из файла
filename = 'ari_results.txt'

r_values = []
ari_values = []

# Регулярное выражение для извлечения числовых значений
pattern = r'r=([\d.]+),\s*ARI=([\d.]+)'

with open(filename, 'r') as file:
    for line in file:
        line = line.strip()
        if line:  # Пропускаем пустые строки
            match = re.search(pattern, line)
            if match:
                r_values.append(float(match.group(1)))
                ari_values.append(float(match.group(2)))

# Создание графика
plt.figure(figsize=(12, 6))
plt.plot(sorted(r_values), ari_values, 'bo-', linewidth=1, markersize=4)
plt.xlabel('r', fontsize=12)
plt.ylabel('ARI', fontsize=12)
plt.title('Зависимость ARI от параметра r', fontsize=14)
plt.grid(True, alpha=0.3)

# Добавление подписей к точкам (необязательно, может быть слишком много точек)
# for i, (r, ari) in enumerate(zip(r_values, ari_values)):
#     if i % 5 == 0:  # Подписываем каждую 5-ю точку, чтобы не загромождать график
#         plt.annotate(f'{ari:.3f}', (r, ari), textcoords="offset points",
#                      xytext=(0,10), ha='center', fontsize=8)

plt.tight_layout()
plt.show()

# Дополнительно: вывод статистики
print(f"Количество точек: {len(r_values)}")
print(f"Диапазон r: от {min(r_values):.3f} до {max(r_values):.3f}")
print(f"Диапазон ARI: от {min(ari_values):.4f} до {max(ari_values):.4f}")
```
