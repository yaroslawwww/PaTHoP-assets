```
import numpy as np
import matplotlib.pyplot as plt

# Загрузка данных из файла
data = np.loadtxt('str_m.txt', delimiter=',')

# Извлечение столбцов
stretch = data[:, 3]  # 4-й столбец (растяжение)
rmse = data[:, 0]     # 1-й столбец (RMSE)
np_vals = data[:, 1]  # 2-й столбец (NP)
mape = data[:, 2]     # 3-й столбец (MAPE)

# Создание графиков
plt.figure(figsize=(15, 10))

# График RMSE vs Растяжение
plt.subplot(3, 1, 1)
plt.scatter(stretch, rmse, s=10, alpha=0.7, color='blue')
plt.xlabel('Степень растяжения')
plt.ylabel('RMSE')
plt.title('Зависимость RMSE от степени растяжения')
plt.grid(True, linestyle='--', alpha=0.7)

# График NP vs Растяжение
plt.subplot(3, 1, 2)
plt.scatter(stretch, np_vals, s=10, alpha=0.7, color='green')
plt.xlabel('Степень растяжения')
plt.ylabel('NP')
plt.title('Зависимость NP от степени растяжения')
plt.grid(True, linestyle='--', alpha=0.7)

# График MAPE vs Растяжение
plt.subplot(3, 1, 3)
plt.scatter(stretch, mape, s=10, alpha=0.7, color='red')
plt.xlabel('Степень растяжения')
plt.ylabel('MAPE')
plt.title('Зависимость MAPE от степени растяжения')
plt.grid(True, linestyle='--', alpha=0.7)

# Настройка общего вида
plt.tight_layout()
plt.savefig('metrics_vs_stretch.png', dpi=300)
plt.show()
```
